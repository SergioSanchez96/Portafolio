```python
from pandas_datareader import data as pdr
import pandas as pd
import numpy as np
import datetime as dt
import yfinance as yfin
yfin.pdr_override()
```

```python
tickers = ['AAPL','^GSPC']
#tickers = ['TSLA','AAPL','MSFT','NKE','AMZN','WMT','WBD','NVDA','GOOGL','F']
start= dt.datetime(2020,1,1) # año, mes,dia
end = dt.datetime(2022,12,31)

data = pdr.get_data_yahoo( tickers,start, end)['Adj Close']
data
```

```python
# variacion porcentual diario
data.pct_change(1)      # 1 por default

# otra forma:
#data/data.shift()-1
```

```python
# retornos logartimicos
retornos_log = np.log(1+data.pct_change())
retornos_log

# otra forma
 #np.log(data/data.shift())
```

```python
# Matriz covarianza anualizada
matriz_cov = retornos_log.cov()*252
matriz_cov
```

```python
# iloc[filas, columna]

# Covarianza de APPLE con el mercado
cov_market = matriz_cov.iloc[0,1]
cov_market
```

```python
# Varianza del mercado
var_market = retornos_log['^GSPC'].var()*252
```

```python
# beta
beta = cov_market/var_market
beta
```

```python
rf = 0.0137
rm = retornos_log['^GSPC'].mean()*252
risk_premium = rm- rf
risk_premium
```

```python
capm = rf + beta*risk_premium
capm
```

```python
# sharpe = (retorno portafolio - tasa libre riesgo)/desviacion estandar portafolio
sharpe = (capm-rf)/(retornos_log['AAPL'].std()*(252**.5))
sharpe
```

```python
print('La beta de', tickers,'es de :',round(beta,3))
print('El retorno CAPM de',tickers,'es de',round(capm*100,3),'%')
print('EL ratio sharpe de ',tickers,'es de ',round(sharpe,3))
```
