## Funciones

```python
def resumen(df):
  porcentaje_vacios = round(df.isnull().mean()*100,2)
  num_unicos = df.nunique()
  resumen = pd.DataFrame({'Variable':df.columns,
                          'Tipo dato':df.dtypes,
                          '% de vacios': porcentaje_vacios,
                          'Unicos': num_unicos,
                          'q25': round(df.quantile(0.25),2),
                          'q50': round(df.quantile(0.5),2),
                          'q75': round(df.quantile(0.75),2),
                          'q100': round(df.quantile(1),2)
                           })
  resumen=resumen.reset_index(drop=True)
  return resumen

def analisis_tabla(df):
  print('Dataset shape :',df.shape)
  tabla = pd.DataFrame(df.dtypes, columns=['dtypes'])
  tabla = tabla.reset_index()
  tabla['Nombre'] = tabla['index']
  tabla = tabla[['Nombre','dtypes']]
  # si necesitas asignar una Serie como columna en un DataFrame utiliza .values
  tabla['% vacios'] = np.round(df.isnull().sum().values*100/df.shape[0],3)
  tabla['Unicos'] = df.nunique().values
  return tabla

def PorcentajeNull(df):
  NulosPorcentaje = round((df.isnull().sum()*100/df.shape[0]).sort_values(ascending=False),2)
  vacios = pd.DataFrame({'Porcentaje Vacios': NulosPorcentaje}).reset_index()
  vacios.columns=['Variable','Porcentaje Vacios']
  return vacios

def datos_unicos(df):
  df_datos_unicos = df.nunique().sort_values().reset_index()
  df_datos_unicos.columns=['Variable','Valores unicos']
  return df_datos_unicos

def grafica_outliers(df,i):
  fig, (ax_box, ax_hist) = plt.subplots(nrows=2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
  sb.boxplot(df[i], orient='h', ax=ax_box)
  sb.kdeplot(df.loc[df['SeriousDlqin2yrs'] == 0, i], color='green', ax=ax_hist, fill=True)
  sb.kdeplot(df.loc[df['SeriousDlqin2yrs'] == 1, i], color='red', ax=ax_hist, fill=True)
  ax_hist.legend(['No Default', 'Default'])
  ax_box.set_title(i)

  ax_box.set(yticks=[])
  sb.despine(ax=ax_hist)
  sb.despine(ax=ax_box, left=True)
  plt.tight_layout()
  plt.show()

def grafica_outliers2(df,i):
  fig, (ax_box, ax_hist) = plt.subplots(nrows=2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
  sb.boxplot(df[i], orient='h', ax=ax_box)
  sb.distplot(df.loc[df['SeriousDlqin2yrs'] == 0, i], color='green', ax=ax_hist)
  sb.distplot(df.loc[df['SeriousDlqin2yrs'] == 1, i], color='red', ax=ax_hist)
  ax_hist.legend(['No Default', 'Default'])
  ax_box.set_title(i)

  ax_box.set(yticks=[])
  sb.despine(ax=ax_hist)
  sb.despine(ax=ax_box, left=True)
  plt.tight_layout()
  plt.show()

def Cuantiles(lista):
    c = [0,1,5,10,20,25,30,40,50,60,70,75,80,90,92.5,95,97.5,99,100]
    matrix = pd.concat([pd.DataFrame(c),pd.DataFrame(np.percentile(lista.dropna(),c))],axis = 1)
    matrix.columns = ["Cuantil","Valor_Cuantil"]
    return(matrix)
```

# Modelo

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score,f1_score

model_df = {}
def probar_modelo(model):
  x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=0, stratify=y)
  model.fit(x_train,y_train)
  y_pred = model.predict(x_test)
  print(f"{model} accuracy es {accuracy_score(y_test,y_pred)}")
  score =cross_val_score(model,x,y, cv=5)
  print(f"{model} promedio de cross validation es {score}") # np.mean(score)
  model_df[model] =  round(np.mean(score)*100,2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
probar_modelo(lr)
```

```python
from sklearn.svm import SVC
svc = SVC()
probar_modelo(svc)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
probar_modelo(dtc)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
probar_modelo(rfc)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
probar_modelo(gbc)
```

# Matriz de confusión

```python
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=0, stratify=y)
lr = LogisticRegression(max_iter=1000)
lr.fit(x,y)
y_pred= lr.predict(x_test)

cm = confusion_matrix(y_test, y_pred)

cm_df = pd.DataFrame(cm, index=['Actual 0', 'Actual 1'], columns=['Predicción 0', 'Predicción 1'])

plt.figure(figsize=(6, 4))
sb.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicción')
plt.ylabel('Valor Real')
plt.title('Matriz de Confusión')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.metrics import f1_score

# Calcular accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calcular precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Calcular recall (sensibilidad)
recall = recall_score(y_test, y_pred)
print("Recall (Sensibilidad):", recall)

# Calcular matriz de confusión para obtener true positives, false positives, etc.
conf_matrix = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = conf_matrix.ravel()

# Calcular sensibilidad manualmente
sensibilidad = tp / (tp + fn)
print("Sensibilidad (Recall):", sensibilidad)

# Calcular F1-score
f1 = f1_score(y_test, y_pred)
print("F1-score:", f1)

# Calcular weighted F1-score
weighted_f1 = f1_score(y_test, y_pred, average='weighted')
print("Weighted F1-score:", weighted_f1)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, auc


y_pred_proba = lr.predict_proba(x_test)[:, 1]

# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calcular el área bajo la curva ROC (AUC-ROC)
auc_score = auc(fpr, tpr)  # otra forma: auc_score = roc_auc_score(y_test, y_pred_proba)

# Crear el gráfico de la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='AUC = %0.2f' % auc_score)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC')
plt.legend(loc="lower right")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, train_test_split

# Cargar los datos de ejemplo
iris = load_iris()
x, y = iris.data, iris.target

# Inicializar el modelo de Regresión Logística
model = LogisticRegression(max_iter=350)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=0, stratify=y)
# Realizar validación cruzada con 5 folds
cv_scores = cross_val_score(model, x_train, y_train, cv=10)

# Calcular los errores de entrenamiento y prueba
train_errors = 1 - cv_scores
test_errors = 1 - cv_scores

# Graficar los errores
plt.plot(np.arange(1, 11), train_errors, label='Training Error')
plt.plot(np.arange(1, 11), test_errors, label='Testing Error')
plt.xlabel('Fold')
plt.ylabel('Error')
plt.title('Training and Testing Error vs. Fold')
plt.legend()
plt.show()
```
