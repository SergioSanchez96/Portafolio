## Variables con NaN

Para aquellos Variables con un porcentaje de vacios menor al 30% se puede utilizar el método de imputación de acuerdo a las caracteristicas y condiciones de cada Variable.

Existen 3 mecanismos de Datos Faltantes:
 * Datos faltantes Completamente Aleatorio(MCAR): los datos que faltan son estrictamente aleatorios y es ajena a los datos
 * Datos faltantes No Aleatorios(MNAR): los datos que faltan siguen un patrón
 * Datos faltantes Aleatorios(MAR): punto intermedio entre los 2 anteriores


Tecnicas para el tratamiento de Vacios :
1. Descarte: elimina los registros.

    - Eliminar toda la fila: puedes perder datos valiosos
	- Eliminar solo la celda


2. Imputación: reemplazas con otros valores calculados. No deberia cambiar la distribución de los datos.
	- Imputación Simple: algoritmo para una sola estimación. Tecnicas:
        * Media o Mediana
		* Imputación por regresión: predice en funcion de las relaciones de los otros datos completos, verificar que haya un tipo de correlacion entre los datos.
		* Imputacion por Hot-Deck: K-NN : el mas usado. Podemos remplazarlo con la media de los N valores más cercanos al valor faltante. Lo podemos usar cuando no haya correlacion entre variables. Desventaja: si hay muchos datos requiere tiempo de computo.

	- Imputación Multiple: en imputacion simple se basan de una sola estimación, aquí se hacen multiples estimaciones y luego se combinan en un unico valor. Metodos:
		* Imputacion multiple con ecuaciones encadenadas( MICE): es mucho mas robusto que los de imputacion simple.Opera bajo el supuesto que los faltantes son MAR. Ejemplo: despues de la media de los vecinos cercanos se realiza varias iteraciones de regresion hasta llegar al mas preciso.

```python
# Se tratarán los variables restantes que tienen un porcentaje de faltantes menor al 20%
NulosPorcentaje2=(data.isnull().sum() * 100 / data.shape[0]).sort_values(ascending=False)
print('Variables con un porcentaje de vacios menor al 30%: ')
NulosPorcentaje2.index[NulosPorcentaje2>0]
```
